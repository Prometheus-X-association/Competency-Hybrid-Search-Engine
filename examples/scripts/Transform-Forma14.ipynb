```python
import os
import json

import pandas as pd
from numpy import nan
```

```python
base_path = "../"
data_path = os.path.join(base_path, "data")
data_path = os.path.join(data_path, "forma14")
```

```python
# Load raw data files
df = pd.read_excel(
    os.path.join(data_path, "Descripteur_Formacode_des_Champs_Semantiques_20241212.xls"),
)
```

```python
df.replace([nan], [None], inplace=True)
```

```python
columns_to_export = [
    "Code du Terme",
    "Descripteur en typo riche",
    "TG (Terme Générique)",
    "Champ sémantique",
    "Synonymes",
    "Synonymes métier",
    "TS (Termes Spécifiques)",
    "TA (Termes Associés)",
    "NA (Note d’Application)",
    "NE (Note d’Explication)",
]
json_data = df[columns_to_export].to_dict("records")
with open(os.path.join(data_path, "cleaned.json"), "w", encoding="utf-8") as f:
    json.dump(json_data, f, indent=2, ensure_ascii=False, separators=(",", ": "))
```
