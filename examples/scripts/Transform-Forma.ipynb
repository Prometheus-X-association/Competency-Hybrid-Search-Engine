```python
import json
import os

import pandas as pd
from numpy import nan
```

```python
base_path = "../"
data_path = os.path.join(base_path, "data")
data_path = os.path.join(data_path, "forma")
```

```python
# NOTE: There was an issue with the original file from the source website.
# In order to fix the encoding issue, we had to open the .xls file in Google Sheets, before downloading it as a .csv file.
# The original file was not properly encoded, which caused issues when reading it with pandas.

# Load raw data files
df = pd.read_csv(
    os.path.join(data_path, "FCod.csv"),
    header=None,
)
df.columns = [
    "code",
    "libelle_maj",
    "title",
    "lheo_key",
    "lheo_balise",
    "famille_code",
    "semantic_field",
    "synonym",
    "synonym_job",
    "category",
    "specific_terms",
    "associated_terms",
    "explication_note",
    "application_note",
    "level",
    "NSF",
    "ROME",
    "previous_version_correspondance",
    "certificateurs",
    "hierarchy_code",
]
```

```python
df.replace([nan], [None], inplace=True)
```

```python
columns_to_export = [
    "code",
    "title",
    "category",
    "NSF",
    "semantic_field",
    "synonym",
    "synonym_job",
    "specific_terms",
    "associated_terms",
    "ROME",
    "explication_note",
    "application_note",
]
json_data = df[columns_to_export].to_dict("records")
with open(os.path.join(data_path, "cleaned.json"), "w", encoding="utf-8") as f:
    json.dump(json_data, f, indent=2, ensure_ascii=False, separators=(",", ": "))
```
